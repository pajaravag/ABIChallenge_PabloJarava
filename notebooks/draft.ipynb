```python
df_train.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis = 1, inplace = True)
```

```python
df_train.dropna(inplace = True)
```

```python
df_train.info()
```

```python
X_train = df_train.drop(['Survived'], axis=1)
y_train = df_train['Survived'].astype(int)
```

```python
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import make_column_transformer, make_column_selector
from sklearn.pipeline import Pipeline
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV

from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn import metrics
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score 

from sklearn.metrics import ConfusionMatrixDisplay
```

```python
pipeline = Pipeline(
    steps=[

        (
            "column_transfomer",
            make_column_transformer(

            (
                    MinMaxScaler(feature_range = (0, 1), ),
                    make_column_selector(dtype_include = 'float64') 
            ),
                    (
                    OneHotEncoder(drop = 'first'),
                    make_column_selector(dtype_include = object) 
                    ),
            remainder = 'passthrough'
            ),
        ),
        (
            "selectKBest",
            SelectKBest(score_func = f_classif),
        ),
        (
            'classifier',
            XGBClassifier(eta = 0.3, n_estimators = 100, objective = 'binary:logistic', eval_metric = 'logloss', use_label_encoder = False,),
        ),
    ],
)

param_grid = {
    #'selectKBest__k': [12, 16, 20, 24],
    'classifier__max_depth': [3, 5, 7],
    'classifier__learning_rate': [0.01, 0.1, 0.3],
    'classifier__lambda': np.linspace(0, 1, 5),
    'classifier__alpha': np.linspace(0, 1, 5),
}


gridSearchCV = GridSearchCV(
    estimator = pipeline,
    param_grid = param_grid,
    cv = 5,
    scoring = 'f1',
    refit = True,
    return_train_score = True,
)

xgb_model = gridSearchCV.fit(X_train, y_train)

#xgb_y_predict = xgb_model.predict(X_test)
#xgb_score = accuracy_score(xgb_y_predict, y_test)
```

```python
pickle.dump(xgb_model, open(os.path.join(parent_directory, 'model', 'model.pkl'), 'wb'))
```
